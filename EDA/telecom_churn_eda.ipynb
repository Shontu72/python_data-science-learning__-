## 5. Target Variable Analysis (Churn) <a id='5'></a>

Churn is our target variable. It follows a **Bernoulli distribution** — its mean equals P(Churn = 1).

```python
# Absolute counts
print("Churn value counts:")
print(df['Churn'].value_counts())
```

```python
# Proportions — the mean of a binary column equals P(1)
print("Churn proportions (Bernoulli distribution):")
print(df['Churn'].value_counts(normalize=True))

print(f"\nChurn rate: {df['Churn'].mean():.2%}")
```

> ⚠️ **~14.5% churn rate** — This is quite high. A churn rate at this level can severely impact revenue and is a strong business signal to investigate further.

This also means the dataset is **imbalanced** (85% non-churn vs 15% churn), which is important to keep in mind for any future ML modeling.

## 7. Filtering & Conditional Queries <a id='7'></a>

Filtering allows us to answer specific business questions.

```python
# Q1: What is the average daytime usage for churned customers?
avg_day_min_churned = df[df['Churn'] == 1]['Total day minutes'].mean()
print(f"Avg day minutes (churned): {avg_day_min_churned:.2f}")
```

```python
# Q2: What is the max international call duration for loyal customers WITHOUT an international plan?
# Note: requires parentheses around each condition when using & or |
max_intl = df[(df['Churn'] == 0) & (df['International plan'] == 'No')]['Total intl minutes'].max()
print(f"Max intl minutes (loyal, no intl plan): {max_intl}")
```

```python
# Q3: Mean numeric features for churned users
df[df['Churn'] == 1].select_dtypes(include=np.number).mean()
```

## 9. Encoding Categorical Variables <a id='9'></a>

We convert Yes/No columns to boolean (True/False) using two approaches:
- `map()` — for a single column
- `replace()` — for applying the same mapping across multiple columns at once

```python
# Mapping dictionary: 'No' -> False, 'Yes': True
d = {'No': False, 'Yes': True}

# Use replace() for both columns for robustness against re-execution
df.replace({'International plan': d, 'Voice mail plan': d}, inplace=True)

df.head()
```

---
## 10. GroupBy Analysis <a id='10'></a>

GroupBy lets us compare subgroups. Syntax: `df.groupby(by=[cols])[cols_to_show].function()`

```python
# Compare call duration stats between churned vs non-churned customers
# percentiles=[] omits quartiles for a cleaner output
df.groupby(['Churn'])[
    ['Total day minutes', 'Total eve minutes', 'Total night minutes']
].describe(percentiles=[])
```

```python
# Compare call duration by International plan status
df.groupby(['International plan'])[
    ['Total day minutes', 'Total eve minutes', 'Total night minutes']
].describe(percentiles=[])
```

```python
# How many churned/loyal customers have an international plan? (absolute counts)
df.groupby(['Churn'])['International plan'].value_counts()
```

```python
# Same, but as proportions within each Churn group
# Key insight: churned customers are ~4x more likely to have an international plan
df.groupby(['Churn'])['International plan'].value_counts(normalize=True)
```

```python
# Detailed look: churned customers WITH an international plan
df[
    (df['Churn'] == 1) & (df['International plan'] == True)
][['Total day minutes', 'Total eve minutes', 'Total night minutes']].describe(percentiles=[])
```

## 11. Summary Tables — Crosstab & Pivot <a id='11'></a>

### 11a. Crosstab
Quick frequency table between two categorical variables.

```python
# Absolute counts: Churn vs International plan-- same work but much more easire than previous groupby function
pd.crosstab(df['Churn'], df['International plan'])
```

```python
# Proportions: Churn vs Voice mail plan
# normalize=True divides every cell by the total count
pd.crosstab(df['Churn'], df['Voice mail plan'], normalize=True)
```

### 11b. Pivot Table
More flexible than crosstab — lets you choose aggregation function and which values to show.

**Syntax:** `df.pivot_table([values], [index], aggfunc='mean')`

```python
# Average number of calls (day/eve/night) broken down by Area code
df.pivot_table(
    ['Total day calls', 'Total eve calls', 'Total night calls'],
    ['Area code'],
    aggfunc='mean'
)
```

```python
# Unique area codes in the dataset
print(f"Area codes: {df['Area code'].unique()}")
```

## 12. Feature Engineering <a id='12'></a>

We create new columns by combining existing ones. This can be useful for downstream modeling.

```python
# New feature: Total charge = sum of all individual charges
df['Total charge'] = (
    df['Total day charge']
    + df['Total eve charge']
    + df['Total night charge']
    + df['Total intl charge']
)

print(f"Total charge range: ${df['Total charge'].min():.2f} — ${df['Total charge'].max():.2f}")
df[['Total day charge', 'Total eve charge', 'Total night charge', 'Total intl charge', 'Total charge']].head()
```

```python
# New feature: Total calls = sum of all call types
df['Total calls'] = (
    df['Total day calls']
    + df['Total eve calls']
    + df['Total night calls']
    + df['Total intl calls']
)

print(f"Total calls range: {df['Total calls'].min()} — {df['Total calls'].max()}")
df[['Total day calls', 'Total eve calls', 'Total night calls', 'Total intl calls', 'Total calls']].head()
```
